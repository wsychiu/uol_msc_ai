## Using the NLTK Library -- BLEU



```python
from nltk.translate.bleu_score import sentence_bleu
candidate = ['this', 'is', 'a', 'perfect', 'test']  #tokenised list
reference = [['this', 'is', 'a', 'test'], ['this', 'is' 'test']]

score = sentence_bleu(reference, candidate)
print(score)
```

## Individual N-Gram Scores Using the NLTK Library
An individual N-gram score is the evaluation of just matching grams of a specific order, such as single words (1-gram) or word pairs (2-gram or bigram).  The weights are specified as a tuple where each index refers to the gram order. To calculate the BLEU score only for 1-gram matches, you can specify a weight of 1 for 1-gram and 0 for 2, 3 and 4 (1, 0, 0, 0).

```python
# two references for one document
from nltk.translate.bleu_score import corpus_bleu

candidate = ['this', 'is', 'test']
references = [['this', 'is', 'a','test'], ['this', 'is' 'test']]

score = sentence_bleu(references, candidate, weights=(1, 0, 0, 0))
print(score)
```

#### Bleu accepts the following input parameters:

- **lowercase:** If True, lowercased BLEU is computed.

- **tokenize:** The tokenizer to use. If None, defaults to language-specific tokenizers with '13a' as the fallback default.

- **smooth_method:** The smoothing method to use ('floor', 'add-k', 'exp', or 'none').

- **smooth_value:** The smoothing value for `floor` and `add-k` methods. `None` falls back to the default value.

- **max_ngram_order:** If given, it overrides the maximum n-gram order (default: 4) when computing precisions.
