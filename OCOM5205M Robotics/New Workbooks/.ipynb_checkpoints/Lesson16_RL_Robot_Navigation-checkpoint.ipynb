## Moving the robot(Actions)
We start by controlling the robot only

```python

class Env_a(Node):

# initialisation--------------------------------------------------------------
    # frequency: how many often (in seconds) the spin_once is invoked, or the publisher is publishing to the /cmd_vel
    def __init__(self, name='Actions_publisher',  frequency=.5, speed=1.0, θspeed=0.3):
        super().__init__(name)
        
        self.speed = speed
        self.θspeed = θspeed
        self.a = -1
        self.robot = Twist()

        # Creating a controller that publishes Twist messages
        #     Message Type = Twist; Topic = '/cmd_vel; Queue Size = 0
        self.controller = self.create_publisher(Twist, '/cmd_vel', 0)

        # Timer that publishes 
        self.timer = self.create_timer(frequency, self.control)

# Control--------------------------------------------------------------
    # move then stop to get a defined action
    def step(self, a):
        print('action is', a)
        self.a = a
        ros.spin_once(self)

        self.a = -1
        ros.spin_once(self)

    # needed to avoid recursion
    def stop(self):
        self.step(-1)
        ros.spin_once(self)

    def control(self):
        # print('moving', self.a)
        speed = θspeed = 0.0
        if self.a==-1: speed = θspeed = 0.0  # print('stop') 
        if self.a==0:  θspeed = self.θspeed  # print('left')  
        if self.a==1:  speed = self.speed    # print('forward')
        if self.a==2:  θspeed = -self.θspeed # print('right') 
        
        self.robot.linear.x = speed
        self.robot.angular.z = θspeed
        self.controller.publish(self.robot)
```

Ok, let us now test our little environment, to do so, open a terminal and launch the simple environment be executing the following command:

<code> ros2 launch turtlebot3_gazebo turtlebot3_simple.launch.py </code>

To make the testing smoother, you can right-click Gazebo and keep the window on top.

```python
def test_robot(env):
    for _ in range(5): env.step(1)
    for _ in range(5): env.step(0)
```

```python
test_robot(Env_a())
```

## Sensing the Environment (States)
Now we add the ability to scan the surroundings to setect walls obsticals etc.
This ability is cricial in order to recognise the states

```python

class Env_sa(Env_a):

# initialisation--------------------------------------------------------------
    def __init__(self, name='Action_Pub_States_Sub'):
        super().__init__(name)
        
        self.range_max = 3.5
        self.range_min = .25  # change as you see fit

        # Create a subscription to LaserScan messages
        #  of the topic /scan
        #  with callback function name self.scan
        #  with Queue size = 0
        self.scanner = self.create_subscription( LaserScan, '/scan', self.scan, 0)
        self.scanner  # prevent unused variable warning


# sensing--------------------------------------------------------------
    
    # change as you see fit
    def scan(self, scans):
        print(f'type(scans): {type(scans)}')
        #print(f'scans.ranges: {scans.ranges}')
        scans = np.array(scans.ranges)
        scans[scans==Inf] = self.range_max
        print('scans = ', scans[:10].round(2))
        self.scans = scans
```

```python
test_robot(Env_sa())
```
