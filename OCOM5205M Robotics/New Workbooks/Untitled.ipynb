```python
Q_table = [[.0,.1,.0,.3],
           [.1,.4,.0,.3],
           [.2,.0,.0,.5],
           [.0,.2,.1,.0],
           [.2,.1,.3,.0],
           [0,0,0,0]]
Q_table = np.array(Q_table)
Q_table
```

```python
for i,s in enumerate(Q_table):
    print(f'S{i}: {s.sum()/4}')
```
