```python
# Dependencies
import nbimporter

import rclpy as ros
from rclpy.node import Node
from sensor_msgs.msg import LaserScan
import numpy as np
from numpy import Inf
from geometry_msgs.msg import Twist
from nav_msgs.msg  import Odometry
from std_srvs.srv import Empty
import random
from gazebo_msgs.srv import SpawnModel
import random
#from math import atan2, pi
import matplotlib.pyplot as plt
```

```python
ros.init()
```

```python
# Creating Environment Class, which achieves the following:
#    1) Publishes robot movement messages
#    2) Subscribes to LaserScan /scan messages
#    3) Translates (2) into state-space
#    4) Returns a reward number based on state

class Env_TurtlebotAssessment2(Node):
    # Initialisation
    def __init__(self, name='ActionPublisher_StateSubscriber', frequency=0.1, speed_linear=1.0, speed_angular=0.25):
        super().__init__(name)

        # Basic default values for movement
        self.speed_linear = speed_linear    # Linear speed in m/s
        self.speed_angular = speed_angular  # Angular speed in rad/s
        self.action = -1                    # Default action set to "stop"
        self.robot = Twist()                # Twist object

        # Robot movement
        #  Creating a publisher for this node to publish Twist messages
        self.controller = self.create_publisher(Twist, '/cmd_vel', 0)
        #  Creating a timer to publish controller messages at set frequency
        self.timer = self.create_timer(frequency, self.control)
    
    # Defining function to make robot take an action
    def step(self, action):
        actions_list = {-1:'stop'
                        , 0:'turn left'
                        , 1:'go forward'
                        , 2:'turn right'}

        self.action = action
        ros.spin_once(self)     # Take action
        print(f'Action taken: {actions_list[self.action]}')

        self.action = -1
        ros.spin_once(self)     # Halt the robot after taking a step (no coasting)
        print(f'Stopped')

    # Defining a stop function
    def stop(self):
        self.step(-1)
        ros.spin_once(self)
    
    # Defining the action-message (i.e. Twist message) to be published
    def control(self):
        speed_linear = speed_angular = 0.0      # Changing default values
        match self.action:
            case -1: speed_linear = speed_angular = 0.0
            case  0: speed_angular = self.speed_angular
            case  1: speed_linear = self.speed_linear
            case  2: speed_angular = -self.speed_angular
            case  _: speed_linear = speed_angular = 0.0     # Unrecognised action == stop
        
        self.robot.linear.x = speed_linear      # Setting linear speed in Twist message
        self.robot.angular.z = speed_angular    # Setting angular speed in Twist message

        self.controller.publish(self.robot)     # Publish Twist message
```

```python
# Defining a test function to test basic functionality
def test_robot(env, steps_to_take):
    for _ in range(steps_to_take): env.step(0)
    for _ in range(steps_to_take): env.step(1)
    for _ in range(steps_to_take): env.step(2)
```

```python
test_robot(Env_TurtlebotAssessment2(frequency=0.05), 10)
```
