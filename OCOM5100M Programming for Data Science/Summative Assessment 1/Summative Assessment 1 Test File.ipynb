```python
def sum_of_numbers (n):
    i = 2
    total = 0
    count = 0
    while count < n:
        for j in range(2, i//2 + 1):
            if i%j == 0:
                break
        else:
            total += i
            count += 1
        i += 1
    return total
```

```python
outputs = []
for n in range(99, 101):
    outputs += (n, sum_of_numbers(n))

outputs
```

```python
def ripple(L):
    M = [L[0]]*len(L)
    print(M)
    for i in range(len(L)-1):
        M[i+1] = L[i+1] + M[i]
    return M
```

```python
def set_analysis(s):
    L = list(s)
    if not L == sorted(L):
        return False
    return set([(L.count(x),x) for x in L ])

m_set = {(3, 'x'), (1, '4'), (3, 'u'), (2, 'e'), (1, 'c')}

set_analysis('4ceeuuuxxx') == m_set
# set_analysis('xxx4uuueec')
```

```python
def number_check(x, y, z):
    if not isinstance(x, int): 
        return False
    return 1 < x < y and x == z and x*y*z == 12 

number_check(2,3,2)
```

```python
def rmtt(s):
    if '-' in s or "'" in s:
        return False
    if not s.islower():
        return False
    s = s.replace('//', '-').title()
    s = s.replace('-And-', " 'n' ")
    return s
# "Steak ‘n’ Chips"

rmtt('steak//and//chips')
```

```python
def xyx( s ):
    return len(s) < 2 or ( s[0] == s[-1] and xyx( s[1: -1] ) ) 

def xyx_key(n):
    s = str(n)
    l = len(s)
    if xyx(s) and l%2==0:
        return int( s[:int(l/2)] )
    return False

# 54631

xyx_key(5463113645)
```
