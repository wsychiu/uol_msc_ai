# Introduction to NumPy

[NumPy](https://numpy.org/doc/stable/user/whatisnumpy.html) is the package for scientific computing in Python. It is a Python library that provides a multidimensional array object, various derived objects (such as matrices), and an assortment of routines for fast operations on arrays. NumPy arrays facilitate advanced mathematical and other types of operations on large numbers of data. Typically, such operations are executed more efficiently and with less code than is possible using Python’s built-in sequences.

## Axes


The axes are the different dimensions of the array. They are zero-indexed. In a 2-D array the vertical axis is 0 and the horizontal is 1. This means we specify the vertical axis first. So a shape [4,3] means 4 numbers vertically and 3 horizontally. 4 rows of 3 numbers each. Many functions have an optional parameter specifying which axis or axes are being affected. For higher-dimensional arrays you could think of it as the 'inner most' dimensions have the highest index. 

Several functions can be applied to different axes. For example, ``sum``.

```python
random_array = np.random.randint(4, 25, size=[3,4])
random_array
```

```python
random_array.sum()
```

```python
random_array.sum(axis = 0) # sum each column
```

```python
random_array.sum(axis = 1) # sum each row
```

### Broadcasting
NumPy is great for avoiding (explicit) loops when you want to do the same thing to every item in your data. 
Adding the same scalar value to every item in an array is one form of broadcasting. More generally numpy can work with arrays of different shapes and interprets the best way to combine them. 

```python
m =  [1,2,3,4,5,6]
n =  [4,5,6,1,2,3]
m*n
```

```python
m =  [1,2,3,4,5,6]
n =  [4,5,6,1,2,3]
Result = []
for i, j in zip (m,n):
    Result.append(i*j)
print("The product of 2 lists is: ", Result)
```

```python
Result = [i * j for i, j in zip(m, n)]  #list comprehension
print(Result)
```

```python
#Array multiplication is done element-wise
x = np.array([[1,2,3], [4,5,6]])
y = np.array([[4,5,6], [1,2,3]])
x.shape, y.shape
```

```python
x*y
```

```python
a = np.array([[1,2,0], [1,0,4], [0,2,5], [6,7,8]])
b = np.array([4,1,3])
a,b
```

```python
a*b
```

```python
a = random_array[0]
a
```

```python
a + 10  #broadcasting
```

Or to target/filter a specific item in each row, we could do the following

```python
a[[0,1,2,3], [4,3,2,0]]
```

```python
a[[0,1,2,3], [4,3,2,0]] += 10
```

```python
a
```

```python
a = np.random.randint(4, 25, size=[4, 5])
a
```

```python
a + np.array([1,2,3,4,5]) # it's not magic
```

``a`` has shape (4,5) and the array ``np.array([1,2,3,4,5])`` has shape (1,5).
Broadcasting works where for each dimension either the size matches or one of the arrays has size 1. In the first case the operation is applied componentwise, in the second case the single value is used repeatedly. 
So adding a scalar to an array is interpreted as adding the scalar value to every element in the array. Adding the array of shape (1, 5) to the array of shape (4, 5) means componentwise adding the values in the single row to each of the four rows. The same principles can be applied across more than 2 dimensions, as the following examples illustrate. 

```python
x = np.array([400])
b = np.array([x, x+100, x+200])
b, b.shape 
```

numpy's _universal functions_ support broadcasting. Mostly arithmetic or similar.

## Example: series approximations

N.B. This example is only fun if you really like maths. 

We can approximate the value of $e^x$ using the series $$\sum_{i = 0}^{\infty} \frac{x^i}{i!}$$


We will use numpy to compare partial sums $\sum_{i = 0}^{n} \frac{x^i}{i!}$ for increasing values of $n$ to the correct value.

The function ``vectorize`` allows a non-numpy function to be applied to all elements in an array, we use this to calculate $n!$, which is used within the partial sum approximation.

```python
import math
import numpy as np

fac = np.vectorize(math.factorial) 

def approx_exp(x, num=10):
    n = np.arange(num)
    return np.sum((x ** n) / fac(n))
```

The following cells break out this function into its parts so we can see what's happening. First decide the range of values we will calculate the partial sum over.

```python
n = np.arange(10)
n
```

Suppose x=2. We calculate $x^i$ and divide this by $i!$ for each value $i$ in the array ``n``.

```python
p = (2 ** n)
p
```

```python
f = fac(n)
f
```

```python
p / f
```

Finally, take the sum.

```python
np.sum(p / f)
```

And this is exactly what the function `approx_exp` above does. 

```python
approx_exp(2, 10) # approximate $e^2$
```

```python
approx_exp(3, 10) # approximate $e^3$
```

The following function allows us to see how the approximation improves as we take the partial sum over more values. 

```python
def pretty_print_exp(x, num):
    print("Actual:", math.e ** x) 

    print("i (terms)\tPartial Sum\tError")

    for i in range(1, num+1):        
        approx_val = approx_exp(x, num=i)
        print(f"{i}\t\t{approx_val:.03f}\t\t{math.e**x - approx_val:.03f}")
```

```python
pretty_print_exp(3, 10)
```

```python
pretty_print_exp(4, 10)
```

We'll try to generalise this so we can pass in different approximations and work on multiple input values (x) at once. We can also see that some calculation is repeated within the python ``for`` loop because calculating a partial sum to $n+1$ necessarily involves also calculating the sum to $n$. We can use numpy to avoid this repeated effort.

```python
input_vals = np.array([3,4])
input_vals
```

```python
n = np.arange(5)
n
```

We want to use these two arrays together but they have different shapes, so we'll need to reshape in order to allow numpy to use broadcasting. 

```python
input_vals = input_vals.reshape(2,1)
n = n.reshape(1,5)
```

```python
input_vals , n
```

Calculate the powers $3^i$ and $4^i$ in one array.

```python
p = np.power(input_vals, n)
p
```

As before, we divide each $x^i$ by $i!$ using the vectorized factorial function. 

```python
d = np.true_divide(p, fac(n))
d
```

And take the sum, note the use of the axis parameter.

```python
np.sum(d, axis = 1)
```

To reduce the work within the python for loop, we can instead use numpy's cumulative sum function.

```python
np.cumsum(d, axis = 1) # get cumulative sum to each position
```

Putting the previous several cells together:

```python
approx_exp = np.cumsum(np.true_divide(np.power(input_vals, n), fac(n)), axis = 1)
```

```python
approx_exp
```

Alternatively as a lambda

```python
approx_exp = lambda x,n: np.cumsum(np.true_divide(np.power(x,n), fac(n)), axis = 1)
```

```python
approx_exp(input_vals, n)
```

Or just as a normal function -- probably more readable and we can put in some validation. 

```python
def approx_exp(x, n):
    assert(x.ndim == 2)
    assert(x.shape[1] == 1)
    assert(n.ndim == 2)
    assert(n.shape[0] == 1)
    
    fac = np.vectorize(math.factorial)
    
    return np.cumsum(np.true_divide(np.power(x,n), fac(n)), axis = 1) 
```

```python
approx_exp(input_vals, n)
```
