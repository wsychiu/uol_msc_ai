## Dunder Methods

[Dunder](https://www.tutorialsteacher.com/python/magic-methods-in-python#:~:text=Python%20%2D%20Magic%20or%20Dunder%20Methods,class%20on%20a%20certain%20action.) or magic methods in Python are the methods having two prefix and suffix underscores in the method name. Dunder here means “Double Under (Underscores)”. These are commonly used for operator overloading. Few examples for magic methods are: __init__, __add__, __len__, __repr__ etc.

Magic methods are not meant to be invoked directly by you, but the invocation happens internally from the class on a certain action. For example, when you add two numbers using the + operator, internally, the __add__() method will be called.


```python
class Employee:
    def __init__(self, Emp_Name, Emp_Salary):
        self.name = Emp_Name
        self.salary = Emp_Salary
    
    def __gt__(self, other):
        return self.salary > other.salary
     
        #__str__ is called when print or str function is invoked on an object
    def __str__(self):
         return 'name='+self.name+' salary=£'+str(self.salary)
    
    def __lt__(self, other):
        return self.salary < other.salary
    
    def __eq__(self, other):
        return self.salary == other.salary
    
    def __iter__(self):
        self.n = self.name
        self.stop = 1
        return self
    
    def __next__(self):  
        if self.n >= self.stop:
            n = self.n
            self.n -= 1
            return n
        else:
            raise StopIteration
    
    def __len__(self):
        return self.name
  
    def __getitem__ (self):
         return self.name, self.salary
    
#   __repr__ returns the string representation of the object 
    def __repr__(self):
        return (f'{self.__class__.__name__}({self.name},{ self.salary})')
```

```python
Emp1 = Employee ('Sam', 1000)
Emp2 = Employee ('Laura', 2000)
Emp3 = Employee ('John', 2500)
Emp4 = Employee ('Emily', 3500)
```
