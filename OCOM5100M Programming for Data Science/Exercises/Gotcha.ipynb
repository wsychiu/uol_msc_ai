## 3

```python
i = 0
while i < 5:
    print(i)
    i += 1
```

Above while loop behaves as expected, and the next cell contains an equivalent for loop. 

```python
for i in range(5):
    print(i)
```

If we want to increase the step size that's easy in the while loop:

```python
i = 0
while i < 5:
    print(i)
    i += 2
```

So why hasn't the step size increased in this for loop?

```python
for i in range(5):
    print(i)
    i += 2
```

## 5

```python
mixed_list = [False, 5, "false", 3.0, False, [], True, 0]

integers_count = 0
booleans_count = 0

for item in mixed_list:
    if isinstance(item, bool):
        booleans_count += 1
    elif isinstance(item, int):
        integers_count += 1
```

```python
integers_count
```

```python
booleans_count
```

Change the order in which we check for Booleans and integers to get a different count.

```python
mixed_list = [False, 5, "false", 3.0, False, [], True, 0]

integers_count = 0
booleans_count = 0

for item in mixed_list:
    if isinstance(item, int):
        integers_count += 1
    elif isinstance(item, bool):
        booleans_count += 1
```

```python
isinstance(False, int)
```

```python
integers_count
```

```python
booleans_count
```

Switch elif for if to change the result again.

```python
mixed_list = [False, 5, "false", 3.0, False, [], True, 0]

integers_count = 0
booleans_count = 0

for item in mixed_list:
    if isinstance(item, bool):
        booleans_count += 1    
    if isinstance(item, int):
        integers_count += 1
```

```python
integers_count
```

```python
booleans_count
```

```python
data = {}
val = []
for i in range(5):
    val.append(i)
    data[i] = val
    
print(data)
```

## 8
Python nan means "not a number", it's a special float value. 

```python
x = float('nan') 
y = float('nan')
```

```python
x is y
```

```python
x == y
```

```python
[x] == [y]
```

Force the two variables to point to the same memory address

```python
x = y
```

Now x and y are identical:

```python
x is y
```

But still they're not the same value.

```python
x == y
```

```python
x == x
```

```python
[x] == [y]
```

A consequence of this - don't use 'nan' as a dictionary key:

```python
my_dict = {}
my_dict[float('nan')] = 'not a number'
```

```python
my_dict[float('nan')]
```

```python
x = float('nan')
my_dict[x] = 'also not a number'
```

```python
my_dict[x]
```
