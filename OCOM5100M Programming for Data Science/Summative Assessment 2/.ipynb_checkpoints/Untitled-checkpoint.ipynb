```python
import pandas as pd
```

```python
df_wellbeing = pd.read_excel('Potential Datasets/personal-well-being-borough.xlsx', sheet_name='Summary - Mean Scores'
                            , header = [0,1], skiprows=[2,3])
df_wellbeing.head()
```

```python
print(df_wellbeing.shape)
df = df_wellbeing.dropna(how='all')
print(df.shape)
df.head()
```

```python
blank_cols = []
for col in df.columns:
    if df[col].isnull().all():
        blank_cols.append(col)

df2 = df.drop(columns=blank_cols)
print(df2.columns)
```

```python
df3 = pd.melt(df2, id_vars=[(' ', 'Code'), (' ', 'Area')], var_name=['Factor', 'Year'], value_name='Median')
df3.head()
```

```python
df3['Adj_Year'] = df3['Year'].apply(lambda x: int('20'+x[-2:]))
df3.head()
```
