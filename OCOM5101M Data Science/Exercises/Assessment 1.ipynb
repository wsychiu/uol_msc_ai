```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
data = np.array([[5,5],[4,5],[4,4],[1,1],[1,2],[2,1]])
print(data)
```

```python
kmeans_movies = KMeans(init ='k-means++', n_clusters = 2)
kmeans_movies.fit(data)
```

```python
kmeans_movies.cluster_centers_
```
