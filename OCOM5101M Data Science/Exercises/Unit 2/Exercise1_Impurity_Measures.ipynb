# Impurity Measures Behaviour
This notebook explore the behaviour of the different components that constitute impurity measures including:
1. Entropy
2. Gini Index
3. Misclassification Error



```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
p = np.linspace(0.01, 0.99, num = 100)
print(max(p))
def log(p):
    return np.log2(p)
```

## Entropy

### - log p

```python
fig, ax1 = plt.subplots(nrows = 1)
ax1.plot( p, -log(p)   , 'b-'  , label = '$-log(p)$')
ax1.legend(loc = 'upper right')
ax1.set_xlabel('p')
ax1.set_ylabel('$-log(p)$')
#plt.savefig('logp.png') # we can use this command to save a figure to the notebook folder
plt.show()
```

### - log (1-p)

```python
fig, ax1 = plt.subplots(nrows = 1)
ax1.plot( p, -log(1-p)   , 'b--'  , label = '$-log(p\')=-log(1-p)$')
ax1.legend(loc = 'upper right')
ax1.set_xlabel('p')
ax1.set_ylabel = '$-log(p\')=log(1-p)$'
plt.show()
```

### -p log(p)

```python
fig, (ax1, ax2) = plt.subplots(nrows = 2, sharex = True)


ax1.plot( p,      p   , 'r-' , label = '$p$')
ax1.plot( p, -log(p)  , 'b-'  , label = '$-log(p)$')
ax1.legend(loc = 'upper right')

ax2.plot(  p, -p*log(p), 'm-', label='$-p*log(p)$')
ax2.legend(loc = 'upper right')
ax2.set_xlabel('p')

plt.subplots_adjust(wspace=0, hspace=0)
#fig.set_size_inches(8, 6) # you can use this to hcnage the figure size 
plt.show()
```

### - (1-p) log(1-p)

```python
fig, (ax1, ax2) = plt.subplots(nrows = 2, sharex = True)

ax1.plot( p,      1-p   , 'r--' ,  label = '$p\'=1-p$')
ax1.plot( p, -log(1-p)  , 'b--'  , label = '$-log(p\')$')
ax1.legend(loc = 'upper right')

ax2.plot(  p, -(1-p)*log(1-p), 'm--', label='$-p\'*log(p\')$')
ax2.legend(loc = 'upper right')
ax2.set_xlabel('$p$')

plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

### Entropy(C1,C0) = -p log p - (1-p) log(1-p)
Where we have 
1. pr(C1) = p
2. pr(C0) = 1-p

```python
fig, ax1 = plt.subplots(nrows = 1, sharex = True)


ax1.plot(  p, -p*log(p), 'm-', label='$Entropy(C1)=-p*log(p)$')
ax1.plot(  p, -(1-p)*log(1-p), 'm--', label='$Entropy(C0)=-p\'*log(p\')$')
ax1.plot(  p, -p*log(p) - (1-p)*log(1-p), 'm-.', label='$Entropy(C1 , C0) = -p*log(p) - p\'*log(p\')$')

ax1.legend(loc = 'best', framealpha=0.0)
ax1.set_xlabel('$p$')

plt.show()
```

This time we will do all components together in one go

```python
fig, (ax1, ax2) = plt.subplots(nrows = 2, sharex = True)

ax1.plot( p,      p   , 'r-' , label = '$p$')
ax1.plot( p, -log(p)  , 'b-'  , label = '$-log(p)$')

ax1.plot( p,      1-p,  'r--',   label = '$p\'=1-p$')
ax1.plot( p, -log(1-p), 'b--',    label = '$-log(p\')$')
ax1.legend(loc = 'upper center', ncol  = 2)

ax2.plot(  p, -p*log(p), 'm-', label='$Entropy(C1)=-p*log(p)$')
ax2.plot(  p, -(1-p)*log(1-p), 'm--', label='$Entropy(C0)=-p\'*log(p\')$')
ax2.plot(  p, -p*log(p) - (1-p)*log(1-p), 'm-.', label='$Entropy(C1 , C0) = -p*log(p) - p\'*log(p\')$')


ax2.legend(loc = 'lower center',framealpha=0.0)
ax2.set_xlabel('$p$')

fig.set_size_inches(6,5)
plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

## Gini Index(C1,C0)

### 1-p

```python
fig, ax1 = plt.subplots(nrows = 1)

ax1.plot( p, 1-p , 'g-' , label = '$1-p$')
ax1.legend(loc = 'upper center')
ax1.set_xlabel('$p$')

plt.show()
```

### p = 1- p'
where we have p'=1-p and p=1-p'

```python
fig, ax1 = plt.subplots(nrows = 1)

ax1.plot( p, p , 'g--' , label = '$1-p\'$')
ax1.legend(loc = 'upper center')
ax1.set_xlabel('$p$')

plt.show()
```

### p*(1-p)

```python
fig, (ax1, ax2) = plt.subplots(nrows = 2, sharex = True)

ax1.plot( p,   p , 'r-' , label = '$  p$')
ax1.plot( p, 1-p , 'g-' , label = '$1-p$')

ax1.legend(loc = 'upper center')

ax2.plot(  p,  p*(1-p), 'y-', label='$p*(1-p)$')
ax2.legend(loc = 'center')
ax2.set_xlabel('$p$')

plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

### p'(1-p')

```python
fig, (ax1, ax2) = plt.subplots(nrows = 2, sharex = True)

ax1.plot( p, 1-p , 'r--' , label = '$  p\'$')
ax1.plot( p,   p , 'g--' , label = '$1-p\'$')

ax1.legend(loc = 'upper center')

ax2.plot(  p,  p*(1-p), 'y--', label='$p\'*(1-p\')$')
ax2.legend(loc = 'center')
ax2.set_xlabel('$p$')

plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

### Gini(C1,C0)
Where we have 
1. pr(C1) = p
2. pr(C0) = 1-p

```python
fig, ax1 = plt.subplots(nrows = 1, sharex = True)

ax1.plot(  p, p*(1-p), 'r-.', label='$Gini(C1)=-p*log(p)$')
ax1.plot(  p, (1-p)*p, 'g--', label='$Gini(C0)=-p\'*log(p\')$')
ax1.plot(  p, p*(1-p) + (1-p)*p, 'y-.+', label='$Gini(C1 , C0) = p*(1-p)+p\'*(1-p\')$')

ax1.legend(loc = 'best', framealpha=0.0)
ax1.set_xlabel('$p$')

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows = 2, sharex = True)

ax1.plot( p,   p , 'r-' , label = '$  p$')
ax1.plot( p, 1-p , 'g-' , label = '$1-p$')

ax1.plot( p, 1-p , 'r--', label = '$p\'$')
ax1.plot( p,   p , 'g--', label = '$1- p\'$')

ax1.legend(loc = 'upper center', ncol  = 2)

ax2.plot(  p, p*(1-p), 'r-.', label='$Gini(C1)=-p*log(p)$')
ax2.plot(  p, (1-p)*p, 'g--', label='$Gini(C0)=-p\'*log(p\')$')
ax2.plot(  p, p*(1-p) + (1-p)*p, 'y-.+', label='$Gini(C1 , C0) = p*(1-p)+p\'*(1-p\')$')
#ax2.plot(  p,  p*(1-p)  + (1-p)*(p), 'y-+', label='$Gini=p*(1-p)+p\'*(1-p\')$')
ax2.legend(loc = 'lower center',framealpha=0.0)
ax2.set_xlabel('$p$')

plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

### All Gini Componenets

```python
fig, (ax1, ax2) = plt.subplots(nrows = 2, ncols=2, sharex = True)

ax1[0].plot( p, 1-p , 'g-' , label = '$1-p$')
ax1[0].plot( p,   p , 'r-' , label = '$  p$')
ax1[0].legend(loc = 'upper center')

ax1[1].plot( p,   p , 'g--', label = '$  p$')
ax1[1].plot( p, 1-p , 'r--', label = '$1-p$')
ax1[1].legend(loc = 'upper center')


ax2[0].plot(  p,  p*(1-p), 'y-', label='$p*(1-p)$')
ax2[0].legend(loc = 'center')
ax2[0].set_xlabel('$p$')

ax2[1].plot(  p,  (1-p)*p, 'y--', label='$(1-p)*p$')
ax2[1].legend(loc = 'center')
ax2[1].set_xlabel('$p$')

ax1[1].set_yticks([])
ax2[1].set_yticks([])


plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

## Comparing Gini and Entropy

```python
fig, (ax1, ax2) = plt.subplots(nrows = 2, sharex = True)


ax1.plot( p, -log(p)  , 'b-' , label = '$-log(p)$')
ax1.plot( p,      p   , 'r-' , label = '$p$')

ax1.plot( p, -log(1-p), 'b--', label = '$-log(1-p)$')
ax1.plot( p,      1-p,  'r--', label = '$1-p$')
ax1.legend(loc = 'upper center', ncol  = 2)


ax2.plot(  p, -p*log(p) - (1-p)*log(1-p), 'm-.',label='$Entropy$')
ax2.plot(  p,  p*(1-p)  + (1-p)*(p),      'y-+', label='$Gini$')
ax2.legend(loc = 'center')
ax2.set_xlabel('$p$')

plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

## Comparing Gini, Entropy and Misclassification Error

**Summary**

As we can see from the previous figures for the different components of the Gini, Entropy and Misclassification:

- Gini and Misclassification have the same min and max which is 0 and 0.5 respectively for a two class problem.
- Entropy has a max of 1.
- Gini and Entropy behave similarly but they differ in range.
- Misclassification error differs to Gini in terms of its increase and decrease.
- Gini increases and decreases quadratically, while Misclassification error increases and decreases linearly. All Three of these impurity measures are symmetrical.
