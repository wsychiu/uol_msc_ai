# Manually creating a Decision Tree

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_vert = pd.read_csv('../datasets/vertebrate.csv')
df_vert['Is Mammal'] = df_vert['Class']=='mammals'
df_vert.drop(columns=['Class'], inplace=True)
df_vert.head()
```

```python
def gini_index(df):
    classes = df['Is Mammal'].value_counts()
    return 1 - sum((classes**2)/sum(classes)**2)
```

```python
def node_gini(df, node_label):
    gini_arr = [gini_index(df[df[node_label]==value]) * 
                len(df[df[node_label]==value])/len(df)
                for value in df[node_label].unique()]
    print(f"{node_label}: {gini_arr}")  
    return sum(gini_arr)
```

```python
def next_node(df):
    node_gains = [[column, gini_index(df_vert) - node_gini(df_vert, column)] 
                 for column in df_vert.columns if column not in ['Name', 'Is Mammal']]
    df_node_gains = pd.DataFrame(node_gains, columns=['Node', 'Gain']).set_index('Node')
    next_node = df_node_gains[df_node_gains['Gain'] == df_node_gains['Gain'].max()].index.astype(str)[0]
    print(next_node)
```

```python
next_node(df_vert)
```

```python
leaf_node_arr = [gini_index(df_vert[df_vert['Gives Birth']==value]) for value in df_vert['Gives Birth'].unique()]
print(leaf_node_arr)
```
