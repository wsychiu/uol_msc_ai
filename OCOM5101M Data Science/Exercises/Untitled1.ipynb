```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(1)

data = np.array([0,30, 40,50, 60,50, 65,55, 80,70])
data = data.reshape(5,2)
print(data.shape)

noise = np.array([1,2, 1,2, 3,3, 2,3, 1,2]).reshape(5,2)
df_noise = pd.DataFrame(noise)
df_data = pd.DataFrame(data)

# creating the data with noise
noisy_data = np.random.normal(data, noise)
df_noisy_data = pd.DataFrame(noisy_data)
```

```python
# Plotting data with noise
plt.plot(df_data[0], df_data[1], 'xk')
plt.plot(df_noisy_data[0], df_noisy_data[1], '.r')

```
