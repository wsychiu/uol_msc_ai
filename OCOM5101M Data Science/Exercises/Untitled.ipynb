```python
# Manually creating the normal distribution array
np.random.seed(1) # re-seeding

mu = 1.5
sigma = 2.2
N = 200

mus = np.ones((N, 1))*mu
sigmas = np.ones((N,1))*sigma
data = np.random.normal(mus,sigmas)

print(data.shape)

np.random.seed(1)
data_efficient = np.random.normal(mu, sigma, N)
print(data_efficient.shape)
```

```python
# Create a set of univariate Gaussian distributions
np.random.seed(1)

mus = np.array([0,30,40,50,60,  25,36,45,80,70])
sigmas = np.array([1,2,1,2,3,  3,2,3,1,2])
N = 2000
M = len(mus)

data = np.random.normal(mus, sigmas, (N, M))
df = pd.DataFrame(data)
df.head()
```

```python
plt.plot(df, '.')
```
