# Joining the tables together to produce one dataset

```python
datasets = [df_demographics, df_claims, df_policy, df_vehicle, df_target]
for dataset in datasets:
    if dataset.index.name != 'CustomerID':
        dataset.set_index('CustomerID', inplace=True)
        print(dataset.index.name)
```

```python
df_joined = df_demographics.join(datasets[1:], how='outer')
print(df_joined.shape)
df_joined.head(2)
```

```python
df_joined.isnull().sum()
```

# Managing Missing Values

## Dropping columns / datapoints that obviously reasonable to drop

```python
# Because "Country" only has one value ("India"), we can reasonably assume
# that the 2 missing values are also = "India", and therefore we can drop
# the column entirely
df_joined.drop(columns='Country', inplace=True)
```

```python
# Because we are only missing 2 of ~29k datapoints (~0.1%), we can
# reasonably drop these two datapoints without affecting the overall
# dataset / analysis
df_joined.dropna(subset='InsuredGender', inplace=True)
print(df_joined.isnull().sum())
print(df_joined.shape)
```

### Sizing our Reported Fraud problem and seeing if our missing values correlate to fraud

```python
print(df_joined['ReportedFraud'].value_counts())
```

```python
# Because the count of NaN of the 'Witnesses' and 'VehicleMake' fields
# are materially smaller than the Reported Fraud problem, it is also
# reasonable to remove datapoints with these null values
df_joined.dropna(subset=['Witnesses', 'VehicleMake'], inplace=True)
print(df_joined.isnull().sum())
print(df_joined.shape)
pd.set_option('display.max_columns', None)
df_joined.head()
```

After initial data cleaning, we have reduced the total size of the dataset by:
 - 1x Column
 - 126x Rows (28836 - 28710), or 0.4%

The remaining null values will need to managed after splitting the dataset into training, validation, and testing sets to avoid data leakage

# Dimensionality Reduction

```python
# Before we edit, let's create a new DataFrame to preserve the joined version
df = df_joined.copy(deep=True)
```

## Feature Creation

```python
# We can get some more insight into the data by looking at combining
# CapitalGains and CapitalLoss to determine Capital Profit
df['CapitalProfit'] = df['CapitalGains'] + df['CapitalLoss']

# We can also get some insight into the time it takes from obtaining
# coverage policy to when the claim is made
df['TimeToClaim'] = pd.to_datetime(df['DateOfIncident']) - pd.to_datetime(df['DateOfPolicyCoverage'])

df.head()
```

## Removing obviously synonymous / collinear or redundant attributes

```python
# Fraud will not be linked to VehicleID
# We can replace CapitalGains and CapitalLoss with CapitalProfit
# We can replace DateOfIncident and DateOfPolicyCoverage with TimeToClaim
# We are also probably not interested in the exact address of the incident
# so we can we use InsuredZipCode as a proxy for this and all other location-date
# We can also use AmountOfTotalClaim rather than the individual break-downs
cols_to_drop = ['VehicleID', 'DateOfIncident', 'DateOfPolicyCoverage', 'IncidentState', 
               'IncidentCity', 'IncidentAddress', 'AmountOfInjuryClaim', 'AmountOfPropertyClaim',
               'AmountOfVehicleDamage', 'CapitalGains', 'CapitalLoss']
df.drop(columns=cols_to_drop, inplace=True)
```

# Making a train-test-split before further data cleaning

The remaining treatment of suspected redundant or synonymous features as well as handling of null or missing values must now take place after the train-test split so that we do not risk data leakage from our training set into our testing set

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df, test_size=0.33, shuffle=True, random_state=42)
print(f'Original Dataset Shape: {df_joined.shape}')
print(f'Training Set Shape: {df_train.shape}')
print(f'Testing Set Shape: {df_test.shape}')
df_train.head()
```
