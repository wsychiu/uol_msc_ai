```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
def describe_df(df):
    print(f'DataFrame Shape: {df.shape}')
    df_describe = df.describe()
    print(df_describe)
    print(df.isnull().sum())
    for col in df.drop(list(df_describe.columns), axis=1).columns:
        print(f'{col} has {df[col].unique().size} unique values: {df[col].unique()}')
```

# Exploring "Train_Demographics.csv"

```python
df_demographics = pd.read_csv('Datasets/Train_Demographics.csv')
df_demographics.head(3)
```

```python
describe_df(df_demographics)
```
