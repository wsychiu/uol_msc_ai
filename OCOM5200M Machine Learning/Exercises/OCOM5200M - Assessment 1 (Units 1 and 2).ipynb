# Question 3
Consider a Bernoulli process with $\mu = 0.3$. Simulate a sequence of 100 outcomes. Interpret the simulated sequence as follows: place a walker at position 0. A '0' is interpreted as a move of size 0, a '1' is interpreted as a move of size 1. Our simulated sequence can thus be interpreted as a random walk. Record the end position of our random walk. Repeat this 10000 times. Histogram these end positions. Your histogram, when appropriately binned, should resemble a Gaussian. Give ﻿mu﻿ and ﻿sigma﻿ of this distribution. If you are confident, you may also use the central limit theorem directly and dispense with simulation.

```python
def bernoulli_simulator(size, mu):
    my_rng = default_rng()
    uniform_vals = my_rng.uniform(size=size)
    results = [1 if x < mu else 0 for x in uniform_vals]
    return results

def bin(n_events, bern_samples, mu):
    return [bernoulli_simulator(bern_samples, mu) for i in range(n_events)]

results = np.array(bin(10000, 100, 0.3))

#print(results[:10])
sum_results = [sum(results[i]) for i in range(results.shape[0])]
plt.hist(sum_results,bins='auto', density=True)

print(np.array(sum_results).mean())
print(np.array(sum_results).std())

plt.show()
```
