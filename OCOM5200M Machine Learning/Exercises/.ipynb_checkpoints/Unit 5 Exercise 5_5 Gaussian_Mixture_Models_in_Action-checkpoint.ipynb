## Introduction

You will be provided with a number or benchmark datasets. The *Old Faithful* datset. This is a famous geyser in Yellowstone park which erupts with astounding regularity. For a number of eruptions two numbers were recorded: the time to next eruption and the duration of the current one. The dataset is available in the file *faithful.dat*.

Also available are two synthetic datasets in file *overlapped.dat* and file *separated.dat*. Each file represents data generated by a mixture of Gaussians ($K=3$). In *separated.dat* the clusters are well separated, on *overlapped.dat* they overlap. The means and covariances that were used to create both files are given below. In this Activity you will experiment with the performance of $K$-means and GMM models. This activity uses an example implementation that follows the formulae presented in Unit 5. This will give you an opportunity to experiment with code. In practice, for realistic machine learning you would use a library implementation, e.g. from *scikit-learn*. We will demonstrate that in a later activity.

### Some helper functions

The following code parses *separated.dat*

```python
import csv

def readdata(fn):
    '''This function quickly parses separated.dat'''
    x=[]
    y=[]
    
    with open(fn) as f:
        lines = f.readlines()
        reader=csv.reader(lines, quotechar='"', delimiter=',',
                     quoting=csv.QUOTE_ALL, skipinitialspace=True)
        next(reader, None)  # skip the headers
        for l in reader:   
            items=l[0].split()
            x.append(float(items[0]))
            y.append(float(items[1]))
    return x, y
  
x, y = readdata('separated.dat')
print(x[:2])
print(y[:2])

plt.plot(x,y,'.')
```

Whilst the following code reads in the *Old faithful* dataset.

```python
import scipy.stats

def processOldFaithFull():
    with open('faithful.dat') as f:
        lines = f.readlines()
        data = np.array([ [float(x.split()[1]),float(x.split()[2])] for x in lines if len(x.split()) == 3 and x.split()[0].isnumeric()])
        zscore = scipy.stats.zscore(data,axis=0)
    return zscore
        
data=processOldFaithFull()
x=data.T[0]
y=data.T[1]

plt.plot(x,y,'r.')
```

## K-means implementation

The following code is a very literal implementation of the $K$-means algorithm. A few steps are demonstrated
on the data from *separated.dat*

```python
def nearestMu(mus,point):
    max_value  = 1e10
    current_k  = 0
    
    for k, mu in enumerate(mus):
        dist = np.linalg.norm(mu-point)
        if dist < max_value:
            current_k = k
            max_value = dist
            
    return current_k

def assignDataPoints(mus,datapoints):
    K=mus.shape[0]
    N=datapoints.shape[1]
    r=np.zeros(N*K)
    r.shape = N,K

    for i,point in enumerate(datapoints.T):
        k=nearestMu(mus,point)
        r[i][k] = 1.
    return r

def inferMus(r, datapoints):
    weighted_sum = r.T.dot(datapoints.T)
    unweighted_sum = r.sum(axis=0)
    unweighted_sum.shape=weighted_sum.shape[0],1
    mus = weighted_sum/unweighted_sum
    return mus

def kMeans(K,datapoints,oldmus):
    r=assignDataPoints(oldmus,datapoints)
    mus=inferMus(r,datapoints)
    return mus, r
    
def sortClusters(datapoints,r):
    results=[]
    
    for i in range(len(r[0])):
        results.append([])
        
    for i, code in enumerate(r):
        ind = np.nonzero(r[i])
        results[ind[0][0]].append(datapoints.T[i])
        
    return results
    
def visualizeClusterData(clusterdata):
    for cluster in clusterdata:
        Ki=np.array(cluster)
        plt.plot(Ki.T[0],Ki.T[1],'.')        
        
    
x, y = readdata('separated.dat')



K=3 # number of clusters
D=2 # dimensionality of data points
oldmus = np.array([[-2.,1.],[2.,5.],[3.,-2.]])

mus1, r1=kMeans(K, np.array([x,y]),oldmus)
clusterdata = sortClusters(np.array([x,y]),r1)


plt.subplot(2, 2, 1)
visualizeClusterData(clusterdata)
plt.plot(oldmus.T[0],oldmus.T[1],'k*')
plt.plot(mus1.T[0],mus1.T[1],'r+')

mus2, r2=kMeans(K, np.array([x,y]),mus1)
clusterdata = sortClusters(np.array([x,y]),r2)

plt.subplot(2, 2, 2)
visualizeClusterData(clusterdata)
plt.plot(mus1.T[0],mus1.T[1],'k*')
plt.plot(mus2.T[0],mus2.T[1],'r+')

mus3, r3=kMeans(K, np.array([x,y]),mus2)
clusterdata = sortClusters(np.array([x,y]),r3)

plt.subplot(2, 2, 3)
visualizeClusterData(clusterdata)
plt.plot(mus2.T[0],mus2.T[1],'k*')
plt.plot(mus3.T[0],mus3.T[1],'r+')

mus4, r4=kMeans(K, np.array([x,y]),mus3)
clusterdata = sortClusters(np.array([x,y]),r3)

plt.subplot(2, 2, 4)
visualizeClusterData(clusterdata)
plt.plot(mus3.T[0],mus3.T[1],'k*')
plt.plot(mus4.T[0],mus4.T[1],'r+')

```

## Gaussian Mixture Model: E/M Algorithm

Below we show an implementation of the E/M Algorithm  and its application on one of the datasets. First we show
initialisation for *separated.dat*

```python
from scipy.stats import multivariate_normal

## E/M algorithm for GMM
x, y = readdata('separated.dat')

def InitializeClusterData(K, D, loc = 0., scale = 20.):
    '''K, the number of clusters, D the dimensionality of the dataset. '''
    pi = np.random.dirichlet(np.ones(K),size=1) # create a random set of weights in vector pi
    mu = rng.normal(loc=loc,scale=5.,size=K*D)  # create a random set of mus in vector mu
    mu.shape = K, D
    sigma = [np.zeros(D*D) for i in range(K)] # create a list of K matrices, initiated with 0, to act as containers for covariance matrix
    for s in sigma: # iterating through each covariance matrix
        s.shape = D, D
        np.fill_diagonal(s,scale) # turn the diagnols into numbers (default at 20)
    
    return pi[0], mu, sigma # pi[0] because pi.shape is (1,K)
    
def gmmIteration(K, datapoints, pi, mu, sigma):
    responsibilities=[] # empty list of responsibilities
    
    ### Expectation step: Creating the responsibilities of the clusters for each datapoint ###
    
    for point in datapoints: # for every [x,y] pair in datapoints
        rs=[] # residuals?
        for p, m, s in zip(pi, mu, sigma): # mind: pi is an array; zipping together the parameters and looping over each set of parameters (aka: cluster)
            rv = multivariate_normal(m,s)  # create a Gaussian distribution based on the provided parameters of a certain cluster
            r=p*rv.pdf(point)  # residual = pi * sampled datapoint's pdf in the Gaussian distribution we just defined
            rs.append(r)  # add this residual to the list of residuals
        resp=np.array(rs)/np.linalg.norm(rs)  # take all of the residuals and divide by the length / magnitude of the vector to get the responsibility
        responsibilities.append(resp)  # add this vector of responsibilities to the list of responibilities

    ### Maximisation step: Determining the next set of means and covariances and weights
    
    resparray=np.array(responsibilities)
    n_k = resparray.sum(axis=0) 
    
    mus=np.zeros(K*D) # Creating a new mean matrix
    mus.shape = K, D
    sigmas = np.zeros(K*D*D) # Creating a new covariance matrix
    sigmas.shape = K, D, D
    pis = np.zeros(K) # Creating a new weight matrix

    for k in range(K): # iterating over each cluster
        for resp, point in zip(resparray.T[k],datapoints): # join the responsibility and the data point together
            mus[k] += resp*point # summing over the product of the responsibility and data point
        mus[k] /= n_k[k] # dividing by the normalisation factor to get our new mean

    # we need a new loop because the new mus are used in the calculation of the new sigmas
    for k in range(K):
        for resp, point in zip(resparray.T[k],datapoints):
            diff = point - mus[k] # calculating the distance between the data point and the newly estimated mean 
            sigmas[k] += resp*np.outer(diff,diff) # summing over the product of the the responsibility, the difference, and the difference transposed to get the new covariance
            pis[k] += resp # summing up all of the responsibilities to get pi (before it gets normalised)
            
        sigmas[k] /= n_k[k] # normalising the covariance matrix by the normalisation factor
    
    N = pis.sum() # calculating the normalisation factor for pi
    pis /= N # normalising pi
           
    return pis, mus, sigmas, resparray

plt.plot(x,y,'k.')
pi, mu, sigma = InitializeClusterData(3,2) # Create a random set of K clusters across D dimensions

def visualizeClusters(mus, sigmas, plotobject = plt, axis = True):
    N    = 200 # number of data points to be created for each Gaussian isotropic plot
    X    = np.linspace(-10, 10, N)
    Y    = np.linspace(-10, 10, N)
    X, Y = np.meshgrid(X, Y)
    pos  = np.dstack((X, Y))
    for m, s in zip(mus,sigmas):
        rv   = multivariate_normal(m, s)
        Z    = rv.pdf(pos)
        plotobject.contour(X, Y, Z)
    if axis == False:
        plotobject.axis('off')


visualizeClusters(mu, sigma)
plt.savefig('start.pdf')
```

Now, we show the results of subsequent steps of the algorithm.

```python
datapoints = np.array([x,y]).T
print(np.array([x,y])[:2])
print(datapoints[:2])
```

```python
nrow=6
ncol=4


fig, axs = plt.subplots(nrow, ncol)
from matplotlib import gridspec


gs = gridspec.GridSpec(nrow, ncol, width_ratios=[1, 1, 1, 1],
         wspace=0.0, hspace=0.0, top=0.95, bottom=0.05, left=0.17, right=0.845) 
datapoints = np.array([x,y]).T
for i in range(nrow):
    for j in range(ncol):
        pi, mu, sigma, resp = gmmIteration(3, datapoints, pi, mu, sigma)
        axs[i][j].set_xlim([-5.,5.])
        axs[i][j].set_ylim([-5.,5.])
        axs[i][j].scatter(x,y,c=resp,s=2)
        visualizeClusters(mu, sigma,axs[i][j],axis=False)

print(mu)
plt.savefig('gmmpractice.pdf')

```

```python
# Running the same algorithm for overlapped.dat
x, y = readdata('overlapped.dat')
datapoints = np.array([x,y]).T
convergence_steps = []

pi, mu, sigma = InitializeClusterData(3,2)

nrow=5
ncol=5
show_ax = False

for i in range(nrow * ncol):
    old_mu = mu
    if i == 0:
        plt.subplot(nrow, ncol, 1)
        plt.plot(x,y,'k.')
        #visualizeClusters(mu, sigma, axis=show_ax)
    else:
        pi, mu, sigma, resp = gmmIteration(3, datapoints, pi, mu, sigma)
        convergence_steps.append(np.linalg.norm(mu - old_mu))
    plt_obj = plt.subplot(nrow, ncol, i+1)
    plt_obj.scatter(x,y,c=resp,s=1)
    visualizeClusters(mu, sigma, plt_obj, axis=show_ax)
plt.tight_layout()
plt.show()
print(convergence_steps)
print(mu)
```
