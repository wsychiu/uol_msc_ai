```python
from pgmpy.models import BayesianNetwork
from pgmpy.factors.discrete.CPD import TabularCPD
G = BayesianNetwork()
```

```python
'''
nodes = ['icyweather', 'battery', 'radio', 'ignition',
         'startermotor', 'petrol', 'starts', 'moves']
G.add_nodes_from(nodes=nodes)

edges = [('icyweather', 'battery'), ('icyweather', 'startermotor'),
         ('battery', 'radio'), ('battery', 'ignition'),
         ('ignition', 'starts'), ('startermotor', 'starts'), ('petrol', 'starts'),
         ('starts', 'moves')]
G.add_edges_from(ebunch=edges)
'''
```

```python
G.add_edge(u='icyweather', v='battery')
G.add_edge(u='icyweather', v='startermotor')
G.add_edge(u='battery', v='radio')
G.add_edge(u='battery', v='ignition')
G.add_edge(u='startermotor', v='starts')
G.add_edge(u='ignition', v='starts')
G.add_edge(u='petrol', v='starts')
G.add_edge(u='starts', v='moves')
```

### Some propsed CPTs for each node are:

- P(IcyWeather) = 0.05
- P(Battery|IcyWeather)=0.95,P(Battery|¬IcyWeather)=0.997.
- P(StarterMotor|IcyWeather) = 0.98, P(StarterMotor|¬IcyWeather) = 0.999.
- P(Radio|Battery)=0.9999, P(Radio|¬Battery)=0.05.
- P(Ignition|Battery)=0.998, P(Ignition|¬Battery)=0.01.
- P(Petrol) = 0.995.
- P(Starts|Ignition,StarterMotor, Petrol) = 0.9999,otherentries 0.0.
- P(Moves|Starts) = 0.998.

```python
cpd_icyweather = TabularCPD('icyweather', 2, [[0.95],[0.05]])

cpd_ignition = TabularCPD('ignition', 2, [[1-0.01, 1-0.998],[0.01, 0.998]],
                          evidence=['battery'], evidence_card=[2])

cpd_radio = TabularCPD('radio', 2, [[1-0.05, 1-0.9999],[0.05, 0.9999]],
                       evidence=['battery'], evidence_card=[2])

cpd_petrol = TabularCPD('petrol', 2, [[1-0.995],[0.995]])

cpd_moves = TabularCPD('moves', 2, [[1, 1-0.998], [0, 0.998]],
                      evidence=['starts'], evidence_card=[2])

cpd_starts = TabularCPD('starts', 2, 
                        [[1,1,1,1,1,1,1,1-0.9999],
                         [0,0,0,0,0,0,0,0.9999]],
                        evidence=['ignition', 'startermotor','petrol'],
                        evidence_card=[2,2,2])

cpd_startermotor = TabularCPD('startermotor', 2, [[1-0.999, 1-0.98],[0.999, 0.98]], 
                              evidence=['icyweather'], evidence_card=[2])

cpd_battery = TabularCPD('battery', 2, [[1-0.997, 1-0.95],[0.997, 0.95]], 
                         evidence=['icyweather'], evidence_card=[2])
G.add_cpds(cpd_icyweather, cpd_ignition, cpd_radio, cpd_petrol,
           cpd_moves, cpd_starts, cpd_startermotor, cpd_battery)
```

```python
print(cpd_battery)
```
