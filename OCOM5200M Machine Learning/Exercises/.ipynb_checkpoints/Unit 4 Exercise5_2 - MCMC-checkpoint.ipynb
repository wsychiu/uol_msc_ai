# Exercise 5.2: Gibbs Sampling

This example again uses the student model but only the variables intll and SAT.
The sampling process generates the next state by sampling one variable given the Markov blanket.
The output shows which state the model is in for each sample.

```python
#This code is modified from the examples at pgmpy.org Copyright (c) 2013-2021 pgmpy
from pgmpy.factors.discrete import TabularCPD
from pgmpy.models import BayesianModel
intel_cpd = TabularCPD('intel', 2, [[0.7], [0.3]])
sat_cpd = TabularCPD('sat', 2, [[0.95, 0.2], [0.05, 0.8]], evidence=['intel'], evidence_card=[2])
student = BayesianModel()
student.add_nodes_from(['intel', 'sat'])
student.add_edge('intel', 'sat')
student.add_cpds(intel_cpd, sat_cpd)
from pgmpy.sampling import GibbsSampling
gibbs_chain = GibbsSampling(student)
sampled = gibbs_chain.sample(size=10)
sampled
```

As in the previous exercise we can sum up the number of times the model is in a particular state and normailse these to estimate the required probability.

```python
#To select a row based on multiple conditions you can use &:
SATandIntel = sampled.loc[(sampled['intel'] == 1) & (sampled['sat']==1)]
SATandIntel
```
