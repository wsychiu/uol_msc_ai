## 2. Understanding the case study

###  Case study analysis (8 marks)

#### 1) Missing data encoding
The dataset contains certain features of a patient's medical history; however, these features are not complete for each patient's record. Where data is missing, the dataset has encoded the missing information with high values (e.g. 99, 99, etc.) and are clearly distinct from non-missing values, which are usually in the low single-digit integer range. Where dates are missing (e.g. the date of death for a patient that did not die), these are coded as '9999-99-99'.

These missing values will be replaced with numpy's `np.nan` values so that the existing encoding is not mistaken for genuine datapoints during analysis and model training.

#### 2) Missing data
Most of the data is either one-hot encoded (e.g. a patient either had a history with Pneumonia or not) or binary (e.g. patient is either male or female). This means that dealing with missingness through averages or interpolation is not appropriate; instead, there will be two different approaches:
1) **Removing records that have missing data**: This is the preferred option provided that it does not remove a material amount of data (e.g. < 10% of the data)
2) **Removing features that have material missingness**: (features that have missing data for > 10% of records)
     - Attempts will be made to try and 'fill-in' any missingness in features that appear to have material amounts of missing data
         - As most of the data is binary ('yes' vs 'no'), filling in missingness with averages may not be appropriate
         - However, we can attempt to see if the missingness is related to any other features
     - Any remaining features with material amounts of missingness in the dataset could indicate that this is a part of a patient's medical history that is not readily available, and so the algorithms will need to learn to classify a patient's risk without this feature.

#### 3) Rescaling and encoding features
Some features (e.g. Age) have units that are much higher than those that are one-hot or binary encoded. To ensure that these features do not skew the analysis, the features will be rescaled to a standard normal distribution. Features that require one-hot or binary encoding will be encoded (e.g. 'DATE_DIED')

#### 4) Creating training and testing data sets
The original dataset will need to be separated into the feature set (X) and the target variable (y) that we are trying to classify the data into.

Additionally, in order to effectively evaluate the performance of the models after training, it is prudent to assess the performance against a dataset that the model has not yet 'seen'. To do this, we will reserve a subset of the data for performance measurement and leverage cross validation to measure model performance.

```python
# We set a missingness threshold of ~10% of records
missingness_threshold = 0.1
cols_to_drop = missingness[missingness >= missingness_threshold].keys().to_list()
print(cols_to_drop)
```

**'PREGNANT'** field is missing ~50% of the data, but would intuitively seem like a part of a patient's medical record that should be simple to fill in.
```python
print('When SEX is Female:')
print(df_raw[df_raw['SEX']==1]['PREGNANT'].value_counts())

print('\nWhen SEX is Male:')
print(df_raw[df_raw['SEX']==2]['PREGNANT'].value_counts())
```

When the COVID-19 patient is male, we can see that there are missing values (value = 97), which should be replaced with "not pregnant" (value = 2).

For female patients, there are a small amount of missing data, which we will need to exclude (provided that the number of records falls below our threshold)

**INTUBED** and **ICU** fields have less obvious correlations with other features, so we will need to show the Pearson correlation coefficients against all other features.

```python
plt.rcParams['figure.figsize'] = [18,3]
df = df_raw.corr(numeric_only=True)[['INTUBED', 'ICU']]
for i in range(len(df.columns)):
    ax = plt.subplot(1,2,i+1)
    ax.bar(df.index, df[df.columns[i]])
    ax.set_xticks(df.index)
    ax.set_xticklabels(df.index, rotation=90, fontsize=8)
    ax.set_title(df.columns[i])
plt.show()

print('Unique values for INTUBED:')
print(df_raw['INTUBED'].value_counts().sort_values())

print('\nUnique values for ICU:')
print(df_raw['ICU'].value_counts().sort_values())

print('\nUnique values for PATIENT_TYPE:')
print(df_raw['PATIENT_TYPE'].value_counts().sort_values())
```

### 3) Rescaling / normalizing features with larger units
We can see from the data plots above that there are three features ('AGE', 'MEDICAL_UNIT', 'CLASIFFICATION_FINAL') that need to be re-scaled as these features have larger numbers due to the nature of their units.
We will rescale the data using `MinMaxScaler` from `sklearn` to bring the range of values down to 0 to 1

```python
scaler = MinMaxScaler()
cols_to_rescale = ['AGE', 'MEDICAL_UNIT', 'CLASIFFICATION_FINAL']
df_data[['Age_n', 'MedicalUnit_n', 'Classification_n']] = scaler.fit_transform(df_data[cols_to_rescale])
df_data.drop(columns=cols_to_rescale, inplace=True)
df_data.head()
```

### 4) Creating X and y (target) variables and splitting into Training and Test sets

The Multi-Layer Perceptron (MLP) is a versatile neural network that can be applied to a range of tasks, including supervised learning, and can be applied to both linear and non-linearly separable classification problems. Such properties are desirable in this case study as we lack any useful pre-existing domain knowledge on drivers and causes of risk of death in COVID-19 patients and so we will not want to restrict the algorithm with constraints on linearity. Unfortunately, the MLP can be susceptible to finding local minima and overfitting, which requires careful selection and verification of hyperparameters (typically through nested cross-validation with grid searching of parameters). The iterative approach to finding and optimising hyperparameters means that it can be computationally expensive.

### b) Setting hyper parameters (rationale) (5 marks)


The following hyper parameters for the `MLPClassifier` were chosen:
 - **hidden_layer_sizes:** Left at default `(100,)` to begin with, so that we can assess how the model performs with different levels of complexity in the optimisation phase
 - **activation:** Left at default (`relu`) as this is appropriate for the classification problem we have; however, we will also assess how the model performs when selecting the `logistic` sigmoid function and `adam` solver
 - **solver:** Here we will select `sgd` as the Stochastic Gradient Descent method can reduce our risk of falling into a local minima
 - **learning_rate:** Here we will select `adaptive` so that the learning rate may be adjusted and the model will have some opportunity to fine-tune to a lower training loss
 - **max_iter:** We will increase this to `500` to increase the chance of convergence

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp = MLPClassifier(solver='sgd', learning_rate='adaptive', max_iter=500)
```

```python
%%time
mlp.fit(X_train, y_train)
y_pred = mlp.predict(X_test)
print(balanced_accuracy_score(y_test, y_pred))
```

Without doing a full assessment of the performance metrics, we suspect that with a balanced accuracy score of ~0.7, we can potentially achieve better results through hyper parameter optimisation.

### c) Optimising hyper parameters (5 marks) 


To find the optimal hyper parameters, we will leverage `GridSearchCV`, which will automate the searching of the parameters across a specified set of options (parameter grid) and also use cross-validation in order to reduce risk of overfitting to a particular set of training data.

```python
mlp_param_grid = {'hidden_layer_sizes' : [(150,150), (200,)]
                     ,'activation' : ['logistic', 'relu']
                     ,'solver' : ['sgd', 'adam']}

n_jobs = 10 # Limiting number of jobs to ensure machine can still be used during optimisation
n_folds = 3 # Limiting number of folds to 3 to reduce compute time
mlp_clf = GridSearchCV(mlp, param_grid=mlp_param_grid, n_jobs=n_jobs, cv=n_folds, error_score=0)
```

### d) Performance metrics for training (5 marks)

```python
def display_model_performance(y_test, y_pred):
    bal_acc_score = balanced_accuracy_score(y_test, y_pred)
    print(f'Balanced accuracy score: {bal_acc_score}')

    precision_recall = precision_recall_fscore_support(y_test, y_pred, average="weighted")
    print(f'Precision and Recall are: {precision_recall}')

    score_f1 = f1_score(y_test, y_pred, pos_label=1, average="weighted")
    print(f'F1 Score is: {score_f1}')

    cm = confusion_matrix(y_test, y_pred, normalize='pred') # Normalize over predicted to readily show false positives and false negatives
    cmatrix = ConfusionMatrixDisplay(confusion_matrix=cm)
    cmatrix.plot()
    plt.show()
```
