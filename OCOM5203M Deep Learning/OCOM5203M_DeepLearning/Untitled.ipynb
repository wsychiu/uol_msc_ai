```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.patches as patches
import scipy.signal
```

```python
# Question 3.1
input = np.array([[7,7,7,4,4],
                    [6,5,7,4,3],
                    [7,8,7,3,3],
                    [1,1,2,2,3],
                    [2,3,3,3,6]])

mask = np.array([[1,0,-1],
                 [1,0,-1],
                 [1,0,-1]])

output = scipy.signal.convolve2d(input, mask, mode='valid')
print(output)
```

```python
answer = ''
for row in output:
    for i,char in enumerate(row):
        answer += str(char)
        if i!=2: answer += ','
    answer += ';'
print(answer)
```
