## Accessing Datasets
Pytorch provides easy access to a number of widely used datasets.  Here we access the CIFAR10 dataset of 60,000 pictures of 10 classes of object. The data is downloaded from the University of Toronto. It will only be downloaded once and is stored locally in the '.data' directory. We download the dataset in two parts, training_data and test_data. Individual images and the corresponding target label can be accessed by indexing. By default, the image is a PIL (Python Imaging Library) image. Since we require the image to be represented as a torch.Tensor, we add a transform argument, which does the conversion on access. Note that this also converts the data values to floats in the interval [0,1]. Note also that the shape of the tensor is 3 X H X W, unlike the images stored as numpy arrays, which are stored as H X W X 3.

```python
import torch
import torchvision
import torchvision.transforms as transforms

import matplotlib.pyplot as plt
import numpy as np
```

Create two instances of the `torchvision.datasets.CIFAR10` class, one for training data and the other for test data.

```python
# temporary fix
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

train_set = torchvision.datasets.CIFAR10(
    root='./data',
    train=True,
    download=True,
    transform=transforms.ToTensor()
)

test_set = torchvision.datasets.CIFAR10(
    root='./data',
    train=False,
    download=True,
    transform=transforms.ToTensor()
)

classes = ("plane", "car", "bird", "cat", "deer", "dog", "frog", "horse", "ship", "truck")
```

```python
Indexing a item in the training set returns a tuple of the image and the label.
```

```python
x = train_set[13255]

print(f"The shape of the chosen image tensor is {x[0].shape} and the label is {x[1]}")
```

To show an image represented as a tensor, we will convert to a numpy array and use imshow in the usual way. This is not straightforward because we have to move the channel from the front to the back, which we do with np.transpose.

```python
def timshow(image):
    nimage = np.transpose(image.numpy(),(1,2,0))
    plt.imshow(nimage)

timshow(x[0])
```

We could use the dataset instances directly to train and test a linear clasaifier, but it is convenient to create an iterator using two instances of the `DataLoader` class to supply fixed-size batches of images (or smaller if any left over at the end).

```python
train_loader = torch.utils.data.DataLoader(
    train_set,
    shuffle=True,
    batch_size=16
)

test_loader = torch.utils.data.DataLoader(
    test_set,
    batch_size=24, # Forward pass only so batch size can be larger
   shuffle=False,
)
```

Use the dataloaders to assemble images into batches. Note that the batch of 16 images are stored as a single 16 X 3 X 32 X 32 tensor

```python
it = iter(train_loader)
images, labels = next(it)

print(images.size())
print(labels)
```

Having created a batch of images and associated labels, we can now display these nicely.

```python
for i in range(16):
    ax = plt.subplot(4, 4, i+1)
    timshow(images[i])
    ax.set_title(f"{labels[i]}: {classes[labels[i]]}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```
